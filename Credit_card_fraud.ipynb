**Under Sampling**

Building a sample dataset contains similar distribution of normal transaction and fraud transaction

```python
legit_sample = legit.sample(n=492)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

**Spliting the Data**

```python
x = new_dataset.drop(columns = 'Class', axis = 1)
y = new_dataset['Class']
```

```python
x.head()
```

```python
y
```

```python
# split the data into taining and testing data
```

```python
 x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
# logistic regression model
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
# model evaluation
# Accuracy score
```

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)

# with the model we predicted output (x_train_prediction) with the help of x_train
# and check the accuracy score with y_train
```

```python
print('Accuracy on training data :', training_data_accuracy)
```

```python
# accuracy score of test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('Accuracy on testing data :', test_data_accuracy)
```

```python
# if the accuracy score of training and testing data have a big differences.
#  It means our model is overfitted or underfitted
```
